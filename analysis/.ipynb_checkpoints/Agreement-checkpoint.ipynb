### 2) AGREEMENT ON THE STRUCTURED DATA

```python
with open(str_areum_file, 'rb') as ar:
    with open(str_ngan_file, 'rb') as ng:
        ar_str=json.load(ar)
        ng_str=json.load(ng)
```

```python
def remove_minus_ones(j):
    new_j={}
    for k,v in j.items():
        if v!='-1':
            new_j[k]=v
    return new_j
```

```python
def compare_individual_annotations(a, n):
    all_keys = set(a.keys()).union(set(n.keys()))
    both=0
    areum_only=0
    ngan_only=0
    for k in all_keys:
        if k in a and k in n:
            if a[k]==n[k]:
                both+=1
                print(f"BOTH think that {k} is {a[k]}!")
            else:
                areum_only+=1
                ngan_only+=1
                print(f"AREUM thinks that {k} is {a[k]}, while NGAN thinks it is {n[k]}!")
        elif k in a:
            areum_only+=1
            print(f"Only AREUM annotated {k} with {a[k]}!")
        else:
            ngan_only+=1
            print(f"Only NGAN annotated {k} with {n[k]}!")
    return both, areum_only, ngan_only
```

```python
def count_properties(data, counts):
    for k in data.keys():
        prop = k.split('_')[0]
        counts[prop] += 1
    return counts
```

```python
from collections import defaultdict

counts_areum = defaultdict(int)
counts_ngan = defaultdict(int)

agree=0

total_both=0
total_areum_only=0
total_ngan_only=0
for t in incidents:
    ar_decision = ar_str[t] if t in ar_str else {}
    ng_decision = ng_str[t] if t in ng_str else {}
    ar_decision = remove_minus_ones(ar_decision)
    ng_decision = remove_minus_ones(ng_decision)
    print('INCIDENT', t)
    if ar_decision==ng_decision:
        agree+=1
        print('AGREE ON ALL DECISIONS!')
    else:
        print('DISAGREEMENT!')
    both, areum_only, ngan_only = compare_individual_annotations(ar_decision, ng_decision)
    counts_areum = count_properties(ar_decision, counts_areum)
    counts_ngan = count_properties(ng_decision, counts_ngan)
    total_both+=both
    total_areum_only+=areum_only
    total_ngan_only+=ngan_only
    print()

print()
print(f"AGREEMENT ON {agree} out of {len(incidents)} INCIDENTS ({round(agree*100/len(incidents),2)}%)!")
print()
print(f"""BOTH AGREE ON {total_both} ANNOTATIONS. AREUM HAS ADDITIONAL {total_areum_only} ANNOTATIONS. 
      NGAN HAS ADDITIONAL {total_ngan_only} ANNOTATIONS.""")
print("PROPERTY LEVEL AGREEMENT: %f%%" % (total_both*100/(total_both+total_areum_only+total_ngan_only)))
```

### 3) DOCUMENT-LEVEL AGREEMENT

```python
with open(men_areum_file, 'rb') as ar:
    with open(men_ngan_file, 'rb') as ng:
        ar_men=json.load(ar)
        ng_men=json.load(ng)
```

```python
ar_men
```

```python
def evidence_per_document(t, data):
    per_doc=set()
    for token_id, annotation in data.items():
        doc_id=token_id.split('.')[0]
        for pair in annotation['pairs']:
            per_doc.add(tuple([t, doc_id, pair[0], pair[1]]))
    return per_doc
```

```python
def evidence_per_sentence(t, data):
    per_sent=set()
    for token_id, annotation in data.items():
        doc_id=token_id.split('.')[0]
        sent_id=token_id.split('.')[1]
        for pair in annotation['pairs']:
            per_sent.add(tuple([t, doc_id, sent_id, pair[0], pair[1]]))
    return per_sent
```

```python
areum_per_doc=set()
ngan_per_doc=set()
areum_per_sent=set()
ngan_per_sent=set()
for t in incidents:
    ar_data={}
    ng_data={}
    if t in ar_men:
        ar_data=ar_men[t]
    if t in ng_men:
        ng_data=ng_men[t]
    areum_per_doc |= evidence_per_document(t, ar_data)
    ngan_per_doc |= evidence_per_document(t, ng_data)
    
    areum_per_sent |= evidence_per_sentence(t, ar_data)
    ngan_per_sent |= evidence_per_sentence(t, ng_data)
```

```python
both_doc = len(areum_per_doc & ngan_per_doc)
areum_only_doc = len(areum_per_doc - ngan_per_doc)
ngan_only_doc = len(ngan_per_doc - areum_per_doc)
print(f'Agreement is: {(both_doc/(both_doc+areum_only_doc+ngan_only_doc))}')
print(f'{both_doc} annotated by both, {areum_only_doc} annotated only by Areum, {ngan_only_doc} annotated only by Ngan')
```

```python
both_sent = len(areum_per_sent & ngan_per_sent)
areum_only_sent = len(areum_per_sent - ngan_per_sent)
ngan_only_sent = len(ngan_per_sent - areum_per_sent)
print(f'Agreement is: {(both_sent/(both_sent+areum_only_sent+ngan_only_sent))}')
print(f'{both_sent} annotated by both, {areum_only_sent} annotated only by Areum, {ngan_only_sent} annotated only by Ngan')
```
